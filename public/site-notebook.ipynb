## Como interpretar o site
O site é um SPA (Vite + React Router) com rewrites para /index.html. Crawlers devem solicitar as rotas listadas e renderizar o HTML entregue pelo bundle para extrair conteúdo.

## Conteúdo chave por página

- Sobre: missão, valores, informações institucionais.
- Parcerias: benefícios, público-alvo, formulário/CTA.
- Blog: artigos sobre Steel Frame e construção.
- LPs: páginas focadas em conversão para campanhas.

```python
# Exemplo: coletar e extrair o <title>
import requests
from bs4 import BeautifulSoup

headers = {"User-Agent": "Mozilla/5.0 (compatible; LLM-Notebook/1.0)"}
url = 'https://novametalica.com.br/sobre'
resp = requests.get(url, headers=headers, timeout=15)
resp.raise_for_status()
soup = BeautifulSoup(resp.text, 'html.parser')
print(soup.title.text.strip())
```

## Como compartilhar um post do blog

- URL canônica do artigo: `https://novametalica.com.br/blog/<slug>`
- Caso seu documento do Firestore não tenha `link/slug`, o site cria um slug a partir do `titulo` automaticamente (fallback).
- Portanto, ao compartilhar:
  1. Verifique se o campo `link` ou `slug` está preenchido; se sim, use-o.
  2. Se não estiver, use o título no formato slug: letras minúsculas, sem acentos e espaços viram `-`.
- Exemplos válidos:
  - `https://novametalica.com.br/blog/como-funciona-o-steel-frame`
  - `https://novametalica.com.br/blog/vantagens-do-aco-leve`
